```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
store_1 = pd.read_csv('store_1_sales.csv')
store_1
```

```python
store_2 = pd.read_csv('store_2_sales.csv')
store_2
```

```python
store_3 = pd.read_csv('store_3_sales.csv')
store_3
```

```python
merged_table = pd.concat([store_1, store_2, store_3], ignore_index=True)
merged_table
```

```python
path_log_file = r"log_file.txt"
```

```python
def process_table(path_log_file):
 global merged_table

 try:
  merged_table.dropna(inplace=True)

  # Запись поврежденных строк в лог-файл
  with open(path_log_file, 'w', encoding='utf-8') as file:
   for index, row in merged_table.iterrows():
    if not row.isnull().any():
      file.write(f"{row.to_string()} \n")

 except Exception as e:
  print(f"Ошибка при обработке таблицы: {e}")

process_table(path_log_file)
```
