```python
import pandas as pd
```

```python
df1 = pd.read_csv('store_1_sales.csv')
df2 = pd.read_csv('store_2_sales.csv')
df3 = pd.read_csv('store_3_sales.csv')
df = pd.concat([df1, df2, df3])
df
```

```python
pd.pivot_table(data=df, values=['Quantity Sold', 'Total Sale'], index=['Category'],
               aggfunc={'Quantity Sold': 'sum', 'Total Sale': ['sum', 'mean']}).to_csv('data.txt', sep='\t')
```
